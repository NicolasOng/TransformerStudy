```python
def generate_from_checkpoint(model, dir, size, dataset, iter, decode):
    model.load_state_dict(torch.load(f'{dir}model_{size}_{dataset}_{iter}.pth', map_location=device))
    print(decode(generate(model, context, 2000, block_size)[0].tolist()))
```

# Sample Generations for the Shakespeare Model

```python
encode, decode, get_batch, vocab_size = load_char_data("./data/tinyshakespeare.txt", block_size, batch_size, device)
```

## tiny

```python
model = DecoderOnlyTransformer(vocab_size, block_size, 128, 6, 2, 0)
model = model.to(device)
```

```python
# Took 20m10.3s on my Laptop
for i in iters:
    print(f"Generating from the model using checkpoint {i}")
    generate_from_checkpoint(model, cp_dir, 'tiny', 'shakespeare', i, decode)
    print("\n"*3)
```

## small

```python
model = DecoderOnlyTransformer(vocab_size, block_size, 256, 6, 4, 0)
model = model.to(device)
```

```python
# Took 46min32.1s on my Laptop
for i in iters:
    print(f"Generating from the model using checkpoint {i}")
    generate_from_checkpoint(model, cp_dir, 'small', 'shakespeare', i, decode)
    print("\n"*3)
```

## base

```python
model = DecoderOnlyTransformer(vocab_size, block_size, 512, 6, 8, 0.1)
model = model.to(device)
```

```python
for i in iters:
    print(f"Generating from the model using checkpoint {i}")
    generate_from_checkpoint(model, cp_dir, 'base', 'shakespeare', i, decode)
    print("\n"*3)
```
